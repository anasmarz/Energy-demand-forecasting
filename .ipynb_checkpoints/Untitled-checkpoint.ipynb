```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import pickle
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('D:\ProjectAnas\Energy-demand-forecasting\daily_electricity_consumption.csv', parse_dates=['Date'], index_col='Date')
data = data.sort_index()
```

```python
# Split the data into training and testing sets
train_size = int(len(data) * 0.8)
train, test = data.iloc[:train_size], data.iloc[train_size:]
y_train = train['Daily_Consumption']
y_test = test['Daily_Consumption']
```

```python
# Define and fit the ARIMA model
model = ARIMA(y_train, order=(5, 1, 0))  # Example order (5, 1, 0)
model_fit = model.fit()
```

```python
# Save the model to disk
with open('energydemand_model.pkl', 'wb') as pkl_file:
    pickle.dump(model_fit, pkl_file)
```

```python
model_fit.summary()
```

```python
# Forecast the future values
y_pred = model_fit.forecast(steps=len(y_test))

# Convert predictions to a DataFrame
y_pred = pd.Series(y_pred, index=y_test.index)

# Plot the results
plt.figure(figsize=(10,5))
plt.plot(y_train, label='Training Data')
plt.plot(y_test, label='Actual Consumption')
plt.plot(y_pred, label='Predicted Consumption')
plt.title('Energy Consumption Forecasting')
plt.xlabel('Time')
plt.ylabel('Consumption')
plt.legend()
plt.show()
```
